## Chapter 11 - Deeper Architectures

This chapter introduces CNN building blocks with tiny examples.
Focus on three ideas: what a convolutional filter does to an image, how padding and stride change output size, and how shapes move through a small CNN.

```python
# !pip -q install torch numpy matplotlib  # install dependencies if needed
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
from torch import nn

plt.style.use("seaborn-v0_8")
%config InlineBackend.figure_format = 'retina'

torch.manual_seed(0)
```

## Convolution (toy example)

We start with a 6x6 toy image containing a bright square in the middle.
The 3x3 filter below is a Laplacian-style edge detector: it compares each center pixel to its neighbors, so flat regions stay quiet while boundaries respond more strongly.

```python
x = torch.zeros(1, 1, 6, 6)
x[:, :, 2:4, 2:4] = 1.0

kernel = torch.tensor(
    [[0.0, 1.0, 0.0],
     [1.0, -4.0, 1.0],
     [0.0, 1.0, 0.0]]
).view(1, 1, 3, 3)

y = F.conv2d(x, kernel, stride=1, padding=0)

print("input shape:", tuple(x.shape))
print("kernel shape:", tuple(kernel.shape))
print("output shape:", tuple(y.shape))
print("kernel:\n", kernel.squeeze())
print("output map:\n", y.squeeze())

fig, axes = plt.subplots(1, 2, figsize=(7, 3))
axes[0].imshow(x.squeeze(), cmap="gray", vmin=0, vmax=1)
axes[0].set_title("Toy image")
axes[1].imshow(y.squeeze(), cmap="coolwarm")
axes[1].set_title("Filter response")
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])
plt.tight_layout()
plt.show()

print("The strongest responses appear near the square's edges because the filter reacts to local changes, not constant regions.")
```

## Padding vs stride (shape intuition)

Padding adds a border around the input, which can preserve spatial size.
Stride changes how far the filter moves each step, so a larger stride produces fewer output positions.
Here the exact values are not important; the goal is to see how the shapes change.

```python
sample = torch.randn(1, 1, 5, 5)
kernel = torch.ones(1, 1, 3, 3)

no_pad = F.conv2d(sample, kernel, stride=1, padding=0)
same_pad = F.conv2d(sample, kernel, stride=1, padding=1)
stride_two = F.conv2d(sample, kernel, stride=2, padding=0)

print("input spatial shape:", tuple(sample.shape[-2:]))
print("stride=1, padding=0 ->", tuple(no_pad.shape[-2:]))
print("stride=1, padding=1 ->", tuple(same_pad.shape[-2:]))
print("stride=2, padding=0 ->", tuple(stride_two.shape[-2:]))
print("Without padding, a 3x3 kernel on a 5x5 input gives a 3x3 output.")
print("Padding=1 keeps the height and width at 5x5.")
print("Stride=2 skips positions, so the output shrinks more aggressively.")
```

## Tiny CNN (shape tracing)

Now trace a mini-batch through a two-block CNN.
Watch the four dimensions `(batch, channels, height, width)`: convolutions change channel count, while max-pooling halves the spatial resolution.

```python
class TinyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 8, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(8, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(16 * 7 * 7, 10),
        )

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)


def trace_shapes(model, x):
    current = x
    print(f"{'input':12s}: {tuple(current.shape)}")

    layer_names = ["conv1", "relu1", "pool1", "conv2", "relu2", "pool2"]
    for name, layer in zip(layer_names, model.features):
        current = layer(current)
        print(f"{name:12s}: {tuple(current.shape)}")

    current = model.classifier[0](current)
    print(f"{'flatten':12s}: {tuple(current.shape)}")

    current = model.classifier[1](current)
    print(f"{'linear':12s}: {tuple(current.shape)}")
    return current


model = TinyCNN()
sample = torch.randn(4, 1, 28, 28)
logits = trace_shapes(model, sample)

print("\nEach MaxPool2d(2) halves height and width: 28 -> 14 -> 7.")
print("The final linear layer maps the flattened 16 x 7 x 7 features to 10 logits.")
```

## Training TinyCNN on Fashion-MNIST

The shape trace above confirmed that TinyCNN handles 28×28 inputs correctly.
Now we actually train it on a small Fashion-MNIST subset to see what depth buys us — accuracy that a simple MLP cannot easily match on image data, because convolutional layers exploit the spatial structure that flat linear layers ignore.

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset

# ── data ──────────────────────────────────────────────────────────────────
transform = transforms.ToTensor()
full_train = datasets.FashionMNIST(root='/tmp/fmnist', train=True,  download=True, transform=transform)
full_val   = datasets.FashionMNIST(root='/tmp/fmnist', train=False, download=True, transform=transform)

train_ds = Subset(full_train, range(5000))
val_ds   = Subset(full_val,   range(1000))

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
val_loader   = DataLoader(val_ds,   batch_size=64, shuffle=False)

# ── training loop ─────────────────────────────────────────────────────────
torch.manual_seed(0)
cnn = TinyCNN()
optimizer = torch.optim.Adam(cnn.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()

cnn_val_acc = []

for epoch in range(1, 11):
    cnn.train()
    running_loss = 0.0
    for xb, yb in train_loader:
        logits = cnn(xb)
        loss = loss_fn(logits, yb)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * len(xb)
    train_loss = running_loss / len(train_ds)

    cnn.eval()
    correct = 0
    with torch.no_grad():
        for xb, yb in val_loader:
            preds = cnn(xb).argmax(dim=1)
            correct += (preds == yb).sum().item()
    val_acc = correct / len(val_ds)
    cnn_val_acc.append(val_acc)
    print(f"Epoch {epoch:2d}  train loss: {train_loss:.4f}  val acc: {val_acc:.3f}")

plt.figure(figsize=(6, 3))
plt.plot(range(1, 11), cnn_val_acc, marker='o', label='TinyCNN val accuracy')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.ylim(0, 1)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

TinyCNN reaches around 84 % validation accuracy in just 10 epochs on this 5 000-sample subset, showing that even a shallow two-block CNN quickly learns spatially meaningful features using weight sharing across the image.

```python
# ── shallow MLP baseline ──────────────────────────────────────────────────
class ShallowMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(28 * 28, 256),
            nn.ReLU(),
            nn.Linear(256, 10),
        )

    def forward(self, x):
        return self.net(x)


torch.manual_seed(0)
mlp = ShallowMLP()
mlp_optimizer = torch.optim.Adam(mlp.parameters(), lr=1e-3)

mlp_val_acc = []

for epoch in range(1, 11):
    mlp.train()
    for xb, yb in train_loader:
        loss = loss_fn(mlp(xb), yb)
        mlp_optimizer.zero_grad()
        loss.backward()
        mlp_optimizer.step()

    mlp.eval()
    correct = 0
    with torch.no_grad():
        for xb, yb in val_loader:
            correct += (mlp(xb).argmax(dim=1) == yb).sum().item()
    mlp_val_acc.append(correct / len(val_ds))

# ── comparison plot ───────────────────────────────────────────────────────
ep = range(1, 11)
plt.figure(figsize=(6, 3))
plt.plot(ep, cnn_val_acc, marker='o', label='TinyCNN')
plt.plot(ep, mlp_val_acc, marker='s', linestyle='--', label='ShallowMLP')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.ylim(0, 1)
plt.legend(frameon=False)
plt.tight_layout()
plt.show()

print(f"Final val accuracy — TinyCNN: {cnn_val_acc[-1]:.3f}  ShallowMLP: {mlp_val_acc[-1]:.3f}")
```

On this small 5 000-sample slice the CNN and MLP are nearly tied (~84 %), which is expected: convolutional weight sharing pays off most when the dataset is large enough for the filters to generalise — a concrete reminder that architecture choice and dataset size are inseparable.

## Chapter 11 Summary

- A convolutional filter responds to local patterns such as edges or changes in intensity.
- Padding changes how borders are handled and can preserve spatial size.
- Stride skips locations and usually reduces the output map size.
- In a CNN, it helps to track shapes as `(batch, channels, height, width)` until the tensor is flattened for classification.
