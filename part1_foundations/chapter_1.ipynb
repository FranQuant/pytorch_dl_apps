```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%config InlineBackend.figure_format = 'retina'
```

## Section 1.1 - Minimal Linear Regression

We start with one simple input feature `X` and one target `y` built from a tiny synthetic dataset.


```python
X = np.array([[0.0], [1.0], [2.0], [3.0]], dtype=float)
y = np.array([0.0, 1.0, 2.1, 2.9], dtype=float)

slope, intercept = np.polyfit(X.ravel(), y, deg=1)
pred = slope * X.ravel() + intercept

print(f"feature shape: {X.shape}")
print(f"target shape: {y.shape}")
print(f"slope: {slope:.3f}")
print(f"intercept: {intercept:.3f}")

for xi, yi, pi in zip(X.ravel(), y, pred):
    print(f"x = {xi:.1f} | y = {yi:.2f} | pred = {pi:.2f}")
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(X.ravel(), y, label='data')
xx = np.linspace(0, 3, 50)
plt.plot(xx, slope * xx + intercept, color='red', label='fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()
```

#### What it teaches

- Features and targets can be represented as simple arrays.
- The basic workflow is `fit`, then `predict`, then inspect the result.
- Even a tiny synthetic example can show how a regression line summarizes a pattern.


## Section 1.2 - Train/Test Split and MAE Evaluation

Next we use slightly noisier synthetic data so we can separate training data from testing data and measure prediction error.


```python
rng = np.random.default_rng(0)
X = np.linspace(0, 4, 20, dtype=float).reshape(-1, 1)
noise = rng.normal(0, 0.05, size=X.shape[0])
y = 0.95 * X.ravel() + 0.1 + noise

indices = rng.permutation(len(X))
split_index = int(0.75 * len(X))
train_idx = indices[:split_index]
test_idx = indices[split_index:]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

slope, intercept = np.polyfit(X_train.ravel(), y_train, deg=1)
y_pred = slope * X_test.ravel() + intercept
mae = np.mean(np.abs(y_test - y_pred))

print(f"train samples: {len(X_train)}")
print(f"test samples: {len(X_test)}")
print(f"test MAE: {mae:.4f}")

for x_value, actual, pred in zip(X_test.ravel()[:5], y_test[:5], y_pred[:5]):
    print(f"x = {x_value:.2f} | actual = {actual:.3f} | predicted = {pred:.3f}")
```

## Chapter 1 Summary

- Set up a problem by separating input features `X` from target values `y`.
- Train a simple model with `fit(...)` and generate outputs with `predict(...)`.
- Use a train/test split to check performance on data the model did not see during training.
- Use a basic error metric such as mean absolute error (MAE) to summarize prediction quality.
